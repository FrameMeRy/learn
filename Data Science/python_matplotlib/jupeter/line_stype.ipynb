```python
import matplotlib.pyplot as plt
```

```python
product2 = [15,30,10,20,60] #ยอดขายสินค้า2
month = [1,2,3,4,5] #เลขเดือน
```

```python
plt.plot(month,product2,color="m", marker="s",linestyle="--")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.plot(month,product2,"ms--")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
