```python
import matplotlib.pyplot as plt
```

```python
student = [10,20,40,30,60] #จำนวนนักเรียน
course = ["PHP","Java","Python","C#","NodeJs"] #ชื่อวิชา
```

```python
color = ["purple","red","cyan","yellow","green"]
exp = [0,0,0,0,0.3]
plt.pie(student,
        labels=course,
        autopct="%.2f%%",
        colors=color,
        shadow=True,
        explode=exp,
        startangle=75)
plt.show()
```
