```python
import matplotlib.pyplot as plt
```

```python
room = ["A","B","C"]
boys = [10,15,5]
girls = [20,7,15]
```

```python
plt.bar(room,boys,label="Boys",color="cyan")
plt.bar(room,girls,bottom=boys,label="Girls",color="pink")
plt.xlabel("Room")
plt.ylabel("Amount")
plt.legend()
```

```python
plt.stackplot(room,boys,girls,labels=["Boys","Grils"],colors=["cyan","pink"])
plt.xlabel("Room")
plt.ylabel("Amount")
plt.legend()
```
