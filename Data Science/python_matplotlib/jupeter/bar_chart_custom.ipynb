```python
import matplotlib.pyplot as plt
```

```python
course = ["Python","Java","C#"]
score = [80,75,50]
c = ["green","orange","blue"]
```

```python
plt.bar(course,score,color = c,width=0.5,edgecolor="black",linewidth=2)
plt.xlabel("Course")
plt.ylabel("score")
plt.title("Score Midterm")
plt.show()
```
