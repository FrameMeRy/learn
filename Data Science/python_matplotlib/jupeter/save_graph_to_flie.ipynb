```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50] #ยอดขายสินค้า1 
product2 = [15,30,10,20,60] #ยอดขายสินค้า2
month = [1,2,3,4,5] #เลขเดือน
```

```python
plt.plot(month,product1)
plt.plot(month,product2)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.savefig("../image/export3.png",transparent=True)
plt.show()
```
