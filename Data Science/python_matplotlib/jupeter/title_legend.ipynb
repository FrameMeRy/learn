```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50] #ยอดขายสินค้า1
product2 = [15,30,10,20,60] #ยอดขายสินค้า2
month = [1,2,3,4,5] #เลขเดือน
```

```python
plt.plot(month,product1,"ms--")
plt.plot(month,product2,"b*--")
data = {"size":20,"color":"red","backgroundcolor":"yellow"}
plt.xlabel("Month",data)
plt.ylabel("Sales",data)
config = {"size":20,"color":"blue"}
plt.title("Top Sale 2023",config,loc="center")
plt.legend(["Mouse","Keyboard"],
           title="Product",
           title_fontsize="large",
           fontsize="large",
           facecolor="pink",
           edgecolor="red",
           borderpad=1,
           labelspacing=0.5
           )
```
