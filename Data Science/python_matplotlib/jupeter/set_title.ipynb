```python
import matplotlib.pyplot as plt
```

```python
product2 = [15,30,10,20,60] #ยอดขายสินค้า2
month = [1,2,3,4,5] #เลขเดือน
```

```python
plt.plot(month,product2,"ms--")
plt.xlabel("Month",size=20,color="red",backgroundcolor="yellow")
plt.ylabel("Sales",size=20,color="blue",backgroundcolor="yellow")
plt.show()
```
